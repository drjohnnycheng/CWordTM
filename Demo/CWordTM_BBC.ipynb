# CWordTM Usage on BBC News

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
%%time
import cwordtm
from cwordtm import meta, util, viz, ta, tm
cwordtm.__version__
```

## Load BBC News

```python
bbc_file = "BBC/BBC News Train.csv"
df = util.load_text(bbc_file, info=True)
```

## Preprocessing Text

```python
%%time
text_list = util.get_text_list(df.iloc[:500], text_col='Text')
text = util.preprocess_text(text_list)
```

## Show Word Cloud

```python
viz.show_wordcloud(text, image='images/disc.png', timing=True)
```

```python
viz.show_wordcloud(text_list, code=2)
```

## Summarize Text

```python
news = df.iloc[:50]['Text']
ta.summary(news, weight=3)   # Summarize the news
```
